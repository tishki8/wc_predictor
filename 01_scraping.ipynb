```python
from datafc import eloratings

world_ranking = eloratings.world_ranking_data()
print(world_ranking.shape)
print(world_ranking.columns.tolist())
world_ranking.head(10)
```

```python
country_codes = eloratings.country_codes_data()
print(country_codes.shape)
country_codes.head(10)
```

```python
elo_named = world_ranking.merge(
    country_codes,
    left_on="country",
    right_on="country_code",
    how="left"
)
elo_named = elo_named[["country_name", "elo", "rank"]]
elo_named.columns = ["team", "elo_rating", "elo_rank"]
elo_named.to_csv("data/raw/elo_ratings.csv", index=False)
print(elo_named.shape)
elo_named.head(10)
```

```python
import requests
from io import StringIO

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
resp = requests.get(wiki_url, headers=headers)
print(resp.status_code)  # should be 200 now

wiki_tables = pd.read_html(StringIO(resp.text))
print(len(wiki_tables))
```

```python
for i, t in enumerate(wiki_tables):
    cols = [str(c) for c in t.columns.tolist()]
    print(i, "| shape:", t.shape, "| cols:", cols[:5])
```

```python
import re

score_pattern = re.compile(r"^\d+(\s*\(\d+\))?\s*[–-]\s*\d+(\s*\(\d+\))?")

match_tables = []
for i, t in enumerate(wiki_tables):
    if t.shape[1] == 3:
        col_home, col_score, col_away = t.columns[:3]
        try:
            val = str(t.iloc[0][col_score]).strip()
        except Exception:
            continue
        if score_pattern.match(val):
            match_tables.append(t)

print(len(match_tables))
```

```python
import re

score_pattern = re.compile(r"^\d+\s*[–-]\s*\d+")

match_rows = []
for i, t in enumerate(wiki_tables):
    if t.shape[1] == 3:
        cols = [str(c).strip() for c in t.columns]
        if score_pattern.match(cols[1]):
            match_rows.append({"home_team": cols[0], "score": cols[1], "away_team": cols[2]})

matches_2026 = pd.DataFrame(match_rows)
matches_2026.to_csv("data/raw/wc2026_matches.csv", index=False)
print(matches_2026.shape)
matches_2026.head(20)
```
